## K-means

### Steps:
  Specify number of clusters *K*
  1. initialize centroids
  2. calculate distance from centroids to all points in datasets
  3. assign datapoints to the closest centroids
  4. compute new controids to the clusters by taking the average of the all data points that belong to each cluster
  5. repeat steps 2-5. until there is no change in centroids
